```python
import numpy as np
import matplotlib.pyplot as plt

# Homework 1a)

a = -101/100
n = np.arange(1, 500, 1)
x = a**n / n 
b = np.cumsum(x)

plt.scatter(n, b)

plt.xlabel("n")
plt.ylabel("$(-101/100)^n/n$")
```

```python
# Homework 1b)

n = np.arange(1, 100, 1)
x = a**n / n 
b = np.cumsum(x)

plt.scatter(n, b)

plt.xlabel("n")
plt.ylabel("$1/n^3$")
```

```python
# Homework 1c)

import scipy.special as sp

n = np.arange(1, 100, 1)
x = np.exp(n) / np.sqrt(sp.factorial(n)) 
b = np.cumsum(x)

plt.scatter(n, b)

plt.xlabel("n")
plt.ylabel("$e^n/\sqrt{n!}$")
```
